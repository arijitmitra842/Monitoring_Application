```python
# Importing libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
# Read the dataset.
housing = pd.read_csv("/kaggle/input/usa-housing-listings/housing.csv")
```

```python
# First five rows of the Housing dataframe.
housing.head()
```

```python
# Calcuting the "Correlation" between Sqfeet and Beds.
housing_sqfeet_beds, p = pearsonr(housing.sqfeet, housing.beds)
print(housing_sqfeet_beds)
```

```python
# Scatter plot to show the relationship between Sqfeet and Beds.
plt.scatter(x = housing.sqfeet, y = housing.beds)
plt.xlabel("Sqfeet")
plt.ylabel("Beds")
plt.show
```
